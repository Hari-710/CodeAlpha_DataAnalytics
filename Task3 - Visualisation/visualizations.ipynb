```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\CodeAlpha\Task3 - Visualisation\books_list.csv")

df.head(5)
```

### Price Analysis

```python
# Distribution of Book Prices

plt.hist(df['Price'], bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of book costs")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Spot outliers

plt.boxplot(
    df['Price'],
    patch_artist=True,
    boxprops=dict(facecolor='skyblue'),
    medianprops=dict(color='black'),
    whiskerprops=dict(color='black'),
    flierprops=dict(markerfacecolor='orange', marker='o', markersize=6, linestyle='None')
)
plt.title("Outlier Detection in Book Prices")
plt.ylabel("Price")
plt.xticks([1], ["Books"])
plt.show()
```

```python
# Top 10 most expensive books

top10 = df.nlargest(10, 'Price')[['Title', 'Price']]
plt.barh(top10['Title'], top10['Price'], color='skyblue', edgecolor='black')
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price")
plt.ylabel("Book Title")
plt.gca().invert_yaxis()
plt.show()
```

### Rating Analysis

```python
rating_counts = df['Rating'].value_counts().sort_index()
plt.bar(rating_counts.index, rating_counts.values, color='skyblue',edgecolor='black')
plt.title("Rating Counts")
plt.xlabel("Rating (Stars)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box-plot price vs Rating

sns.boxplot(x='Rating',y='Price',data=df, palette="Blues")
plt.title("Book Prices by Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Price")
plt.show()
```

### Availability Insights

```python
# Bar chart - Availability by Rating

avg_price_rating = df.groupby('Rating')['Price'].mean()

plt.bar(avg_price_rating.index, avg_price_rating.values, color='lightgreen',edgecolor='black')
plt.title("Average Price by Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Average Price")
plt.show()
```
